```python
while True:
    cv2.imshow('result',img)
    if cv2.waitKey(1)==27:
        break;
cv2.destroyAllWindows()
```

```python
haar_data=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_alt2.xml')
```

```python
haar_data.detectMultiScale(img)
```

```python
while True:
    faces=haar_data.detectMultiScale(img)
    for x,y,w,z in faces:
        cv2.rectangle(img,(x,y),(x+w,y+w),(255,0,255),4)
    cv2.imshow('result',img)
    if cv2.waitKey(1)==27:
        break;
cv2.destroyAllWindows()
```

```python
capture=cv2.VideoCapture(0)
data1=[]
while True:
    flag, img = capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,z in faces:
            cv2.rectangle(img,(x,y),(x+w,y+w),(255,0,255),4)
            face=img[y:y+z,x:x+w, :]
            face=cv2.resize(face,(50,50))
            print(len(data1))
            if len(data1)<800:
                data1.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(1)==27 or len(data1)>=200:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
capture=cv2.VideoCapture(0)
data2=[]
while True:
    flag, img = capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,z in faces:
            cv2.rectangle(img,(x,y),(x+w,y+w),(255,0,255),4)
            face=img[y:y+z,x:x+w, :]
            face=cv2.resize(face,(50,50))
            print(len(data2))
            if len(data2)<400:
                data2.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(1)==27 or len(data2)>=200:
            break
capture.release()
cv2.destroyAllWindows()
```
